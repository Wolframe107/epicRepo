```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import pylab
```

```python

img = cv2.imread('stockholm.jpg')

plt.imshow(img[:,:,2], cmap="gray")
def quantize(img):
    img = img.astype('float32')
    img = img/32
    img = img.round()
    img *= 32
    img = img.astype('uint8')
    return img

test = quantize(img[:,:,2])

#print(test)
#plt.imshow(test, cmap="gray")
print(np.unique(test))

rows = []

for i in range(len(test)):
    row = ""

    for j in range(len(test[i])):
        if test[i][j] == 0:
            row += " "

        elif test[i][j] == 32:
            row +=  "."

        elif test[i][j] == 64:
            row +=  "-"

        elif test[i][j] == 96:
            row +=  "*"

        elif test[i][j] == 128:
            row +=  "/"

        elif test[i][j] == 160:
            row +=  "("

        elif test[i][j] == 192:
            row +=  "&"

        elif test[i][j] == 224:
            row +=  "#"

    rows.append(row)

complete = "\n".join(rows)

with open('readme.txt', 'w') as f:
    f.write(complete)

```
